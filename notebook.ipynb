```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visualization style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

```python
# Step 2: Load the GDP data
gdp_df = pd.read_excel('GDP_Turkey.xlsx')

# Display basic information about the GDP data
print("=== GDP Data Overview ===")
print(f"Shape: {gdp_df.shape}")
print("\nFirst 5 rows:")
display(gdp_df.head())

# Check for missing values
print("\nMissing values:")
print(gdp_df.isnull().sum())

# Display data types
print("\nData types:")
print(gdp_df.dtypes)
```

```python
# Step 3: Load the election data for all three years
election_2015_df = pd.read_excel('2015_Elections.xlsx')
election_2018_df = pd.read_excel('2018_Elections.xlsx')
election_2023_df = pd.read_excel('2023_Elections.xlsx')

# Function to display basic information about each election dataset
def explore_election_data(df, year):
    print(f"\n=== {year} Election Data Overview ===")
    print(f"Shape: {df.shape}")
    print("\nFirst 5 rows:")
    display(df.head())
    print("\nColumns:")
    print(df.columns.tolist())
    print("\nMissing values:")
    print(df.isnull().sum())

# Explore each election dataset
explore_election_data(election_2015_df, 2015)
explore_election_data(election_2018_df, 2018)
explore_election_data(election_2023_df, 2023)
```

```python
# Step 4: Group parties into blocs as specified

# Function to create bloc groups based on election year
def create_bloc_groups(df, year):
    # Create a copy to avoid modifying the original dataframe
    result_df = df.copy()

    if year == 2015:
        # Conservative Right Bloc: AK PARTİ, MHP
        conservative_cols = ['AK PARTİ', 'MHP']
        # Centrist Bloc: CHP
        centrist_cols = ['CHP']
        # Left Bloc: HDP
        left_cols = ['HDP']

    elif year == 2018:
        # Conservative Right Bloc: AK Parti, MHP
        conservative_cols = ['AK Parti', 'MHP']
        # Centrist Bloc: CHP, İYİ, Saadet Partisi
        centrist_cols = ['CHP', 'İYİ', 'Saadet Partisi']
        # Left Bloc: HDP
        left_cols = ['HDP']

    elif year == 2023:
        # Conservative Right Bloc: AK Parti, BBP, MHP, YRP
        conservative_cols = ['AK Parti', 'BBP', 'MHP', 'YRP']
        # Centrist Bloc: CHP, İYİ
        centrist_cols = ['CHP', 'İYİ']
        # Left Bloc: Yeşil Sol, TİP
        left_cols = ['Yeşil Sol', 'TİP']

    # Check if all columns exist in the dataframe
    for col_list in [conservative_cols, centrist_cols, left_cols]:
        for col in col_list:
            if col not in df.columns:
                print(f"Warning: Column '{col}' not found in {year} election data.")

    # Create new columns for each bloc
    try:
        result_df['Conservative_Right_Bloc'] = result_df[conservative_cols].sum(axis=1)
        result_df['Centrist_Bloc'] = result_df[centrist_cols].sum(axis=1)
        result_df['Left_Bloc'] = result_df[left_cols].sum(axis=1)

        # If there's a "Diğer" column, keep it as is
        if 'Diğer' in result_df.columns:
            result_df['Other'] = result_df['Diğer']

        # Select only the necessary columns
        # Assuming there's a city/province column - adapt this as needed
        id_cols = [col for col in result_df.columns if col in ['ŞEHİR', 'city', 'province', 'şehir', 'il adı']]
        result_df = result_df[id_cols + ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc', 'Other']]

    except Exception as e:
        print(f"Error creating blocs for {year}: {e}")

    return result_df

# Create bloc groups for each election year
election_2015_blocs = create_bloc_groups(election_2015_df, 2015)
election_2018_blocs = create_bloc_groups(election_2018_df, 2018)
election_2023_blocs = create_bloc_groups(election_2023_df, 2023)

# Display the resulting dataframes
print("\n=== 2015 Election Blocs ===")
display(election_2015_blocs.head())

print("\n=== 2018 Election Blocs ===")
display(election_2018_blocs.head())

print("\n=== 2023 Election Blocs ===")
display(election_2023_blocs.head())
```

```python
# 4.5: Make 2015's City names start with capital letters but all other letters are lower case. Also Sort the rows alphabetically.

election_2015_blocs["ŞEHİR"] = election_2015_blocs["ŞEHİR"].str.title()

election_2015_blocs = election_2015_blocs.sort_values(by="ŞEHİR")


# Also change ŞEHİR to CITY
election_2015_blocs.rename(columns={'ŞEHİR': 'CITY'}, inplace=True)
election_2018_blocs.rename(columns={'ŞEHİR': 'CITY'}, inplace=True)
election_2023_blocs.rename(columns={'ŞEHİR': 'CITY'}, inplace=True)


display(election_2015_blocs.head())


```

```python
# Step 5: Visualize the bloc distribution for each election year. KEEP IN MIND THAT THIS DATA IS BASED ON PERCENTAGES. NOT THE NUMBER OF VOTES.

# Function to create bloc distribution pie charts
def plot_bloc_distribution(df, year):
    # Calculate the mean percentage for each bloc across all cities
    bloc_sums = {
        'Conservative Right': df['Conservative_Right_Bloc'].mean(),
        'Centrist': df['Centrist_Bloc'].mean(),
        'Left': df['Left_Bloc'].mean()
    }

    # Add 'Other' if present
    if 'Other' in df.columns:
        bloc_sums['Other'] = df['Other'].mean()

    # Create a pie chart
    plt.figure(figsize=(5, 5))
    colors = ['#f39c12', '#e74c3c', '#2ecc71', '#3498db']

    plt.pie(
        bloc_sums.values(),
        labels=bloc_sums.keys(),
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        shadow=True
    )
    plt.title(f'Average Political Bloc Distribution - {year} Elections', fontsize=15)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

# Create pie charts for each election year
plot_bloc_distribution(election_2015_blocs, 2015)
plot_bloc_distribution(election_2018_blocs, 2018)
plot_bloc_distribution(election_2023_blocs, 2023)
```

```python
# Step 6: Analyze changes in bloc support over time

# First, we need to ensure we have a common city/province column across all datasets
# This code assumes the city/province column is identified automatically
# You may need to adjust this based on the actual column names in your data

def get_city_column(df):
    for col in df.columns:
        if col.lower() in ['il', 'city', 'province', 'şehir', 'il adı']:
            return col
    return None

# Get city columns for each dataframe
city_col_2015 = get_city_column(election_2015_blocs)
city_col_2018 = get_city_column(election_2018_blocs)
city_col_2023 = get_city_column(election_2023_blocs)

# Check if city columns were found
if None in [city_col_2015, city_col_2018, city_col_2023]:
    print("Warning: Could not identify city/province columns in all datasets.")
    # We'll need to manually specify the columns
    # For example:
    # city_col_2015 = 'Il'  # Adjust based on your data
    # city_col_2018 = 'Il'
    # city_col_2023 = 'Il'

# Create a new dataframe to track changes over time
# We'll use the city names from the 2015 dataset as the base
temporal_df = pd.DataFrame({
    'City': election_2015_blocs[city_col_2015]
})

# Add data from each election year
bloc_types = ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']

for bloc in bloc_types:
    temporal_df[f'{bloc}_2015'] = election_2015_blocs[bloc].values
    temporal_df[f'{bloc}_2018'] = election_2018_blocs[bloc].values
    temporal_df[f'{bloc}_2023'] = election_2023_blocs[bloc].values

# Display the temporal dataframe
print("\n=== Temporal Bloc Changes ===")
display(temporal_df.head())

# Create a line plot showing the average support for each bloc over time
plt.figure(figsize=(12, 7))

# Calculate mean values for each bloc for each year
for bloc in bloc_types:
    years = [2015, 2018, 2023]
    values = [
        temporal_df[f'{bloc}_2015'].mean(),
        temporal_df[f'{bloc}_2018'].mean(),
        temporal_df[f'{bloc}_2023'].mean()
    ]

    # Plot the line for this bloc
    plt.plot(years, values, marker='o', linewidth=3, label=bloc.replace('_', ' '))

plt.title('Average Political Bloc Support Over Time (2015-2023)', fontsize=16)
plt.xlabel('Election Year', fontsize=14)
plt.ylabel('Average Support (%)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.xticks(years)
plt.tight_layout()
plt.show()
```

```python
# Step 8: Prepare data for correlation analysis

# Function to prepare data for a specific year
def prepare_correlation_data(election_df, year, gdp_df):
    # Get city columns for both dataframes
    election_city_col = get_city_column(election_df)
    gdp_city_col = get_city_column(gdp_df)

    if election_city_col is None or gdp_city_col is None:
        print(f"Warning: Could not identify city columns for {year} correlation analysis.")
        return None

    # Create a new dataframe with election bloc data
    corr_df = election_df[[election_city_col, 'Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']].copy()

    # Identify GDP column for the specific year
    gdp_year_col = None
    for col in gdp_df.columns:
        if str(year) in str(col):
            gdp_year_col = col
            break

    if gdp_year_col is None:
        print(f"Warning: Could not find GDP data for year {year}.")
        # Try to use closest available year
        available_years = [col for col in gdp_df.columns if isinstance(col, (int, str)) and str(col).isdigit()]
        if available_years:
            closest_year = min(available_years, key=lambda x: abs(int(str(x)) - year))
            gdp_year_col = closest_year
            print(f"Using GDP data from {closest_year} instead.")
        else:
            print("No suitable GDP data found.")
            return None

    # Merge election data with GDP data
    gdp_subset = gdp_df[[gdp_city_col, gdp_year_col]].copy()
    gdp_subset.columns = [gdp_city_col, 'GDP']

    merged_df = pd.merge(
        corr_df,
        gdp_subset,
        left_on=election_city_col,
        right_on=gdp_city_col,
        how='inner'
    )

    return merged_df

# Prepare correlation data for each election year
corr_2015_df = prepare_correlation_data(election_2015_blocs, 2015, gdp_df)
corr_2018_df = prepare_correlation_data(election_2018_blocs, 2018, gdp_df)
corr_2023_df = prepare_correlation_data(election_2023_blocs, 2023, gdp_df)

# Display the correlation dataframes
for year, df in zip([2015, 2018, 2023], [corr_2015_df, corr_2018_df, corr_2023_df]):
    if df is not None:
        print(f"\n=== {year} Correlation Data ===")
        display(df.head())
    else:
        print(f"\n=== {year} Correlation Data could not be prepared ===")
```

```python
# Step 9: Perform correlation analysis

# Function to analyze correlation for a specific year
def analyze_correlation(df, year):
    if df is None:
        print(f"Cannot analyze correlation for {year} - data not available.")
        return

    print(f"\n=== Correlation Analysis for {year} ===")

    # Calculate correlation matrix
    corr_matrix = df[['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc', 'GDP']].corr()
    print("Correlation Matrix:")
    display(corr_matrix)

    # Create heatmap of correlation matrix
    plt.figure(figsize=(6, 6))
    sns.heatmap(
        corr_matrix,
        annot=True,
        cmap='coolwarm',
        vmin=-1,
        vmax=1,
        linewidths=.5
    )
    plt.title(f'Correlation Between GDP and Political Bloc Support - {year}', fontsize=15)
    plt.tight_layout()
    plt.show()

    # Create scatter plots for each bloc vs GDP
    fig, axes = plt.subplots(3, 1, figsize=(8, 12))
    blocs = ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']

    for i, bloc in enumerate(blocs):
        sns.regplot(
            x='GDP',
            y=bloc,
            data=df,
            ax=axes[i],
            scatter_kws={'alpha': 0.6},
            line_kws={'color': 'red'}
        )
        axes[i].set_title(f'{bloc.replace("_", " ")} vs GDP - {year}', fontsize=14)
        axes[i].set_xlabel('GDP (USD)', fontsize=12)
        axes[i].set_ylabel(f'{bloc.replace("_", " ")} Support (%)', fontsize=12)
        axes[i].grid(True, alpha=0.3)

        # Add correlation coefficient to the plot
        corr = df[['GDP', bloc]].corr().iloc[0, 1]
        axes[i].annotate(
            f'Correlation: {corr:.3f}',
            xy=(0.05, 0.95),
            xycoords='axes fraction',
            fontsize=12,
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", alpha=0.8)
        )

    plt.tight_layout()
    plt.show()

# Analyze correlation for each election year
analyze_correlation(corr_2015_df, 2015)
analyze_correlation(corr_2018_df, 2018)
analyze_correlation(corr_2023_df, 2023)
```

```python
# Step 10: Perform hypothesis testing

from scipy import stats

def perform_hypothesis_tests(df, year):
    if df is None:
        print(f"Cannot perform hypothesis tests for {year} - data not available.")
        return

    print(f"\n=== Hypothesis Tests for {year} ===")

    # For each bloc, test the hypothesis that there is no correlation between GDP and bloc support
    blocs = ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']

    for bloc in blocs:
        # Calculate Pearson correlation coefficient and p-value
        corr, p_value = stats.pearsonr(df['GDP'], df[bloc])

        print(f"\nPearson Correlation Test for {bloc.replace('_', ' ')} and GDP:")
        print(f"Correlation coefficient: {corr:.3f}")
        print(f"p-value: {p_value:.4f}")

        # Interpret results
        alpha = 0.05
        if p_value < alpha:
            print(f"Result: Reject the null hypothesis. There is a statistically significant correlation between GDP and {bloc.replace('_', ' ')} support.")
        else:
            print(f"Result: Fail to reject the null hypothesis. There is not enough evidence to conclude a significant correlation between GDP and {bloc.replace('_', ' ')} support.")

        # Calculate the strength of the relationship
        if abs(corr) < 0.3:
            strength = "weak"
        elif abs(corr) < 0.7:
            strength = "moderate"
        else:
            strength = "strong"

        print(f"The relationship is {strength} ({corr:.3f}).")

# Perform hypothesis tests for each election year
perform_hypothesis_tests(corr_2015_df, 2015)
perform_hypothesis_tests(corr_2018_df, 2018)
perform_hypothesis_tests(corr_2023_df, 2023)
```
