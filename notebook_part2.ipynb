```python
# Step 4: Group parties into blocs as specified

# Function to create bloc groups based on election year
def create_bloc_groups(df, year):
    # Create a copy to avoid modifying the original dataframe
    result_df = df.copy()

    if year == 2015:
        # Conservative Right Bloc: AK PARTİ, MHP
        conservative_cols = ['AK PARTİ', 'MHP']
        # Centrist Bloc: CHP
        centrist_cols = ['CHP']
        # Left Bloc: HDP
        left_cols = ['HDP']

    elif year == 2018:
        # Conservative Right Bloc: AK Parti, MHP
        conservative_cols = ['AK Parti', 'MHP']
        # Centrist Bloc: CHP, İYİ, Saadet Partisi
        centrist_cols = ['CHP', 'İYİ', 'Saadet Partisi']
        # Left Bloc: HDP
        left_cols = ['HDP']

    elif year == 2023:
        # Conservative Right Bloc: AK Parti, BBP, MHP, YRP
        conservative_cols = ['AK Parti', 'BBP', 'MHP', 'YRP']
        # Centrist Bloc: CHP, İYİ
        centrist_cols = ['CHP', 'İYİ']
        # Left Bloc: Yeşil Sol, TİP
        left_cols = ['Yeşil Sol', 'TİP']

    # Check if all columns exist in the dataframe
    for col_list in [conservative_cols, centrist_cols, left_cols]:
        for col in col_list:
            if col not in df.columns:
                print(f"Warning: Column '{col}' not found in {year} election data.")

    # Create new columns for each bloc
    try:
        result_df['Conservative_Right_Bloc'] = result_df[conservative_cols].sum(axis=1)
        result_df['Centrist_Bloc'] = result_df[centrist_cols].sum(axis=1)
        result_df['Left_Bloc'] = result_df[left_cols].sum(axis=1)

        # If there's a "Diğer" column, keep it as is
        if 'Diğer' in result_df.columns:
            result_df['Other'] = result_df['Diğer']

        # Select only the necessary columns
        # Assuming there's a city/province column - adapt this as needed
        id_cols = [col for col in result_df.columns if col in ['ŞEHİR', 'city', 'province', 'şehir', 'il adı']]
        result_df = result_df[id_cols + ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc', 'Other']]

    except Exception as e:
        print(f"Error creating blocs for {year}: {e}")

    return result_df

# Create bloc groups for each election year
election_2015_blocs = create_bloc_groups(election_2015_df, 2015)
election_2018_blocs = create_bloc_groups(election_2018_df, 2018)
election_2023_blocs = create_bloc_groups(election_2023_df, 2023)

# Display the resulting dataframes
print("\n=== 2015 Election Blocs ===")
display(election_2015_blocs.head())

print("\n=== 2018 Election Blocs ===")
display(election_2018_blocs.head())

print("\n=== 2023 Election Blocs ===")
display(election_2023_blocs.head())
```

```python
# Step 8: Prepare data for correlation analysis

# Function to prepare data for a specific year
def prepare_correlation_data(election_df, year, gdp_df):
    # Get city columns for both dataframes
    election_city_col = get_city_column(election_df)
    gdp_city_col = get_city_column(gdp_df)

    # Create a new dataframe with election bloc data
    corr_df = election_df[[election_city_col, 'Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']].copy()


    # Identify GDP column for the specific year
    gdp_year_col = None
    for col in gdp_df.columns:
        if str(year) in str(col):
            gdp_year_col = col
            break

    # Merge election data with GDP data
    gdp_subset = gdp_df[[gdp_city_col, gdp_year_col]].copy()
    gdp_subset.columns = [gdp_city_col, 'GDP']

    merged_df = pd.merge(
        corr_df,
        gdp_subset,
        left_on=election_city_col,
        right_on=gdp_city_col,
        how='inner'
    )


    return merged_df

# Prepare correlation data for each election year
corr_2015_df = prepare_correlation_data(election_2015_blocs, 2015, gdp_df)
corr_2018_df = prepare_correlation_data(election_2018_blocs, 2018, gdp_df)
corr_2023_df = prepare_correlation_data(election_2023_blocs, 2023, gdp_df)

# Display the correlation dataframes
for year, df in zip([2015, 2018, 2023], [corr_2015_df, corr_2018_df, corr_2023_df]):
    if df is not None:
        print(f"\n=== {year} Correlation Data ===")
        display(df.head())
    else:
        print(f"\n=== {year} Correlation Data could not be prepared ===")
```

```python
# Step 9: Perform correlation analysis

# Function to analyze correlation for a specific year
def analyze_correlation(df, year):
    if df is None:
        print(f"Cannot analyze correlation for {year} - data not available.")
        return

    print(f"\n=== Correlation Analysis for {year} ===")

    # Calculate correlation matrix
    corr_matrix = df[['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc', 'GDP']].corr()
    print("Correlation Matrix:")
    display(corr_matrix)

    # Create heatmap of correlation matrix
    plt.figure(figsize=(6, 6))
    sns.heatmap(
        corr_matrix,
        annot=True,
        cmap='coolwarm',
        vmin=-1,
        vmax=1,
        linewidths=.5
    )
    plt.title(f'Correlation Between GDP and Political Bloc Support - {year}', fontsize=15)
    plt.tight_layout()
    plt.show()

    # Create scatter plots for each bloc vs GDP
    fig, axes = plt.subplots(3, 1, figsize=(8, 12))
    blocs = ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']

    for i, bloc in enumerate(blocs):
        sns.regplot(
            x='GDP',
            y=bloc,
            data=df,
            ax=axes[i],
            scatter_kws={'alpha': 0.6},
            line_kws={'color': 'red'}
        )
        axes[i].set_title(f'{bloc.replace("_", " ")} vs GDP - {year}', fontsize=14)
        axes[i].set_xlabel('GDP (USD)', fontsize=12)
        axes[i].set_ylabel(f'{bloc.replace("_", " ")} Support (%)', fontsize=12)
        axes[i].grid(True, alpha=0.3)

        # Add correlation coefficient to the plot
        corr = df[['GDP', bloc]].corr().iloc[0, 1]
        axes[i].annotate(
            f'Correlation: {corr:.3f}',
            xy=(0.05, 0.95),
            xycoords='axes fraction',
            fontsize=12,
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", alpha=0.8)
        )

    plt.tight_layout()
    plt.show()

# Analyze correlation for each election year
analyze_correlation(corr_2015_df, 2015)
analyze_correlation(corr_2018_df, 2018)
analyze_correlation(corr_2023_df, 2023)
```

```python
# Step 10: Perform hypothesis testing

from scipy import stats

def perform_hypothesis_tests(df, year):
    if df is None:
        print(f"Cannot perform hypothesis tests for {year} - data not available.")
        return

    print(f"\n=== Hypothesis Tests for {year} ===")

    # For each bloc, test the hypothesis that there is no correlation between GDP and bloc support
    blocs = ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']

    for bloc in blocs:
        # Calculate Pearson correlation coefficient and p-value
        corr, p_value = stats.pearsonr(df['GDP'], df[bloc])

        print(f"\nPearson Correlation Test for {bloc.replace('_', ' ')} and GDP:")
        print(f"Correlation coefficient: {corr:.3f}")
        print(f"p-value: {p_value:.4f}")

        # Interpret results
        alpha = 0.05
        if p_value < alpha:
            print(f"Result: Reject the null hypothesis. There is a statistically significant correlation between GDP and {bloc.replace('_', ' ')} support.")
        else:
            print(f"Result: Fail to reject the null hypothesis. There is not enough evidence to conclude a significant correlation between GDP and {bloc.replace('_', ' ')} support.")

        # Calculate the strength of the relationship
        if abs(corr) < 0.3:
            strength = "weak"
        elif abs(corr) < 0.7:
            strength = "moderate"
        else:
            strength = "strong"

        print(f"The relationship is {strength} ({corr:.3f}).")

# Perform hypothesis tests for each election year
perform_hypothesis_tests(corr_2015_df, 2015)
perform_hypothesis_tests(corr_2018_df, 2018)
perform_hypothesis_tests(corr_2023_df, 2023)
```

```python
# Step 12: Create comprehensive dataset with all years and engineered features

def create_enhanced_dataset():
    # Start with the temporal dataframe we created earlier
    enhanced_df = temporal_df.copy()

    # Add GDP data for each year
    # First, merge GDP data
    gdp_cols_to_merge = []
    for year in [2015, 2018, 2023]:
        for col in gdp_df.columns:
            if str(year) in str(col):
                gdp_cols_to_merge.append((year, col))
                break

    # Get the city column from GDP data
    gdp_city_col = get_city_column(gdp_df)

    # Merge GDP data for each year
    for year, gdp_col in gdp_cols_to_merge:
        gdp_subset = gdp_df[[gdp_city_col, gdp_col]].copy()
        gdp_subset.columns = [gdp_city_col, f'GDP_{year}']

        enhanced_df = pd.merge(
            enhanced_df,
            gdp_subset,
            left_on='City',
            right_on=gdp_city_col,
            how='inner'
        )
        # Remove duplicate city column
        if gdp_city_col in enhanced_df.columns and gdp_city_col != 'City':
            enhanced_df = enhanced_df.drop(gdp_city_col, axis=1)

    print(f"Base dataset created with shape: {enhanced_df.shape}")
    return enhanced_df

# Create the enhanced dataset
enhanced_df = create_enhanced_dataset()
display(enhanced_df.head())
```

```python
# Step 13: Feature Engineering

def engineer_features(df):
    feature_df = df.copy()

    # 1. GDP GROWTH FEATURES (Economic momentum)
    if 'GDP_2015' in feature_df.columns and 'GDP_2018' in feature_df.columns:
        feature_df['GDP_Growth_2015_2018'] = ((feature_df['GDP_2018'] - feature_df['GDP_2015']) / feature_df['GDP_2015']) * 100

    if 'GDP_2018' in feature_df.columns and 'GDP_2023' in feature_df.columns:
        feature_df['GDP_Growth_2018_2023'] = ((feature_df['GDP_2023'] - feature_df['GDP_2018']) / feature_df['GDP_2018']) * 100

    # 2. POLITICAL MOMENTUM FEATURES (Bloc support changes)
    blocs = ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']

    for bloc in blocs:
        # Change from 2015 to 2018
        feature_df[f'{bloc}_Change_2015_2018'] = feature_df[f'{bloc}_2018'] - feature_df[f'{bloc}_2015']
        # Change from 2018 to 2023
        feature_df[f'{bloc}_Change_2018_2023'] = feature_df[f'{bloc}_2023'] - feature_df[f'{bloc}_2018']

        # Historical average support
        feature_df[f'{bloc}_Historical_Avg'] = (feature_df[f'{bloc}_2015'] + feature_df[f'{bloc}_2018']) / 2

    # 3. DOMINANCE FEATURES (Which bloc was historically strongest)
    for year in [2015, 2018]:
        year_blocs = [f'{bloc}_{year}' for bloc in blocs]
        feature_df[f'Dominant_Bloc_{year}'] = feature_df[year_blocs].idxmax(axis=1)
        feature_df[f'Dominant_Bloc_{year}'] = feature_df[f'Dominant_Bloc_{year}'].str.replace(f'_{year}', '')

        # Dominance strength (how much the winning bloc won by)
        max_vals = feature_df[year_blocs].max(axis=1)
        second_max_vals = feature_df[year_blocs].apply(lambda x: x.nlargest(2).iloc[-1], axis=1)
        feature_df[f'Dominance_Margin_{year}'] = max_vals - second_max_vals

    # 4. ECONOMIC TIER FEATURES
    gdp_cols = [col for col in feature_df.columns if 'GDP_' in col and col not in ['GDP_Growth_2015_2018', 'GDP_Growth_2018_2023']]

    if gdp_cols:
        # Average GDP across all years
        feature_df['GDP_Average'] = feature_df[gdp_cols].mean(axis=1)

        # Economic tier (High/Medium/Low based on GDP)
        feature_df['Economic_Tier'] = pd.qcut(feature_df['GDP_Average'],
                                             q=3,
                                             labels=['Low', 'Medium', 'High'])

    print(f"Feature engineering completed. New shape: {feature_df.shape}")
    print(f"New features created: {feature_df.shape[1] - df.shape[1]}")

    return feature_df

# Apply feature engineering
enhanced_df = engineer_features(enhanced_df)

# Display new features
print("\n=== New Engineered Features Sample ===")
new_feature_cols = [col for col in enhanced_df.columns if any(keyword in col for keyword in
                   ['Growth', 'Change', 'Dominant', 'Dominance', 'Tier', 'Average', 'Historical'])]
display(enhanced_df[['City'] + new_feature_cols].head())
```

```python
# Step 14: Apply feature transformations as required

def apply_transformations(df):
    transformed_df = df.copy()

    # 1. LOG TRANSFORMATION for GDP features (to handle skewness)
    gdp_features = [col for col in df.columns if 'GDP_' in col and 'Growth' not in col]

    for col in gdp_features:
        # Add small constant to avoid log(0)
        transformed_df[f'Log_{col}'] = np.log(transformed_df[col] + 1)

    # 2. POLYNOMIAL FEATURES for non-linear relationships
    # We'll create squared terms for key features
    key_features = ['GDP_Average', 'GDP_Growth_2015_2018', 'GDP_Growth_2018_2023']

    for feature in key_features:
        if feature in transformed_df.columns:
            transformed_df[f'{feature}_Squared'] = transformed_df[feature] ** 2

    # 3. INTERACTION FEATURES between GDP and political trends
    if 'GDP_Average' in transformed_df.columns:
        for bloc in ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']:
            if f'{bloc}_Historical_Avg' in transformed_df.columns:
                transformed_df[f'GDP_x_{bloc}_Interaction'] = (
                    transformed_df['GDP_Average'] * transformed_df[f'{bloc}_Historical_Avg']
                )

    # 4. STANDARDIZATION (will be applied later in ML pipeline)
    # We'll note which features need standardization
    numeric_features = transformed_df.select_dtypes(include=[np.number]).columns.tolist()
    print(f"Numeric features for standardization: {len(numeric_features)} features")

    print(f"Transformations completed. Final shape: {transformed_df.shape}")

    return transformed_df, numeric_features

# Apply transformations
final_df, numeric_features = apply_transformations(enhanced_df)

# Display transformation results
print("\n=== Feature Transformation Summary ===")
transformation_features = [col for col in final_df.columns if any(keyword in col for keyword in
                          ['Log_', '_Squared', '_Interaction'])]
print(f"Transformation features created: {len(transformation_features)}")
print("Sample transformation features:", transformation_features[:5])

# Show final dataset structure
print(f"\nFinal dataset shape: {final_df.shape}")
print(f"Total features created through engineering: {final_df.shape[1] - temporal_df.shape[1]}")
```

```python
# Step 15: Prepare data for machine learning

def prepare_ml_data(df):
    # Training features: Only use 2015 and 2018 data + engineered features from them
    training_features = []

    # Add 2015 and 2018 bloc results
    for year in [2015, 2018]:
        for bloc in ['Conservative_Right_Bloc', 'Centrist_Bloc', 'Left_Bloc']:
            training_features.append(f'{bloc}_{year}')

    # Add GDP data for 2015 and 2018
    for year in [2015, 2018]:
        training_features.append(f'GDP_{year}')

    # Add engineered features that don't use 2023 data
    safe_engineered_features = [
        'GDP_Growth_2015_2018',
        'Conservative_Right_Bloc_Change_2015_2018',
        'Centrist_Bloc_Change_2015_2018',
        'Left_Bloc_Change_2015_2018',
        'Conservative_Right_Bloc_Historical_Avg',
        'Centrist_Bloc_Historical_Avg',
        'Left_Bloc_Historical_Avg',
        'Dominance_Margin_2015',
        'Dominance_Margin_2018',
        'Log_GDP_2015',
        'Log_GDP_2018',
        'GDP_Average'
    ]

    # Only include features that actually exist
    available_safe_features = [f for f in safe_engineered_features if f in df.columns]
    training_features.extend(available_safe_features)

    # Make sure all training features exist
    final_training_features = [f for f in training_features if f in df.columns]

    # Training data (X)
    X = df[final_training_features].fillna(df[final_training_features].mean())

    # Target variables (y) - 2023 results we want to predict
    y = df[['Conservative_Right_Bloc_2023', 'Centrist_Bloc_2023', 'Left_Bloc_2023']]

    print(f"Training features (X) shape: {X.shape}")
    print(f"Target variables (y) shape: {y.shape}")
    print(f"Number of training features: {len(final_training_features)}")

    print("\nTraining features used:")
    for i, feature in enumerate(final_training_features):
        print(f"  {i+1}. {feature}")

    print(f"\nPredicting 2023 results using 2015-2018 historical data")

    return X, y, final_training_features

# Prepare the data
X_train, y_target, feature_names = prepare_ml_data(final_df)
```

```python
# Step 16: Build regression models

def build_regression_models(X, y, feature_names):
    models = {
        'Linear Regression': LinearRegression(),
        'Ridge Regression': Ridge(alpha=1.0),
        'Lasso Regression': Lasso(alpha=0.1),
        'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    }

    results = {}
    all_predictions = {}

    for name, model in models.items():
        print(f"\nTraining {name}...")

        # Create pipeline with standardization for some models
        if name in ['Ridge Regression', 'Lasso Regression']:
            pipeline = Pipeline([
                ('scaler', StandardScaler()),
                ('model', model)
            ])
        else:
            pipeline = Pipeline([
                ('model', model)
            ])

        # Train and evaluate for each bloc
        bloc_results = {}
        bloc_predictions = {}

        for i, bloc in enumerate(['Conservative_Right', 'Centrist', 'Left']):
            target = y.iloc[:, i]

            # Cross-validation to simulate train/test split
            cv_scores = cross_val_score(pipeline, X, target, cv=5, scoring='r2')

            # Fit on all available data to make final predictions
            pipeline.fit(X, target)
            final_predictions = pipeline.predict(X)

            # Calculate prediction error
            actual_values = target.values
            mae = np.mean(np.abs(actual_values - final_predictions))
            rmse = np.sqrt(np.mean((actual_values - final_predictions) ** 2))

            bloc_results[bloc] = {
                'cv_r2_mean': cv_scores.mean(),
                'cv_r2_std': cv_scores.std(),
                'mae': mae,
                'rmse': rmse,
                'pipeline': pipeline
            }

            bloc_predictions[bloc] = final_predictions

            print(f"  {bloc} Bloc:")
            print(f"    Cross-val R²: {cv_scores.mean():.3f} (±{cv_scores.std():.3f})")
            print(f"    Prediction MAE: {mae:.2f}%")
            print(f"    Prediction RMSE: {rmse:.2f}%")

        results[name] = bloc_results
        all_predictions[name] = bloc_predictions

    return results, all_predictions

# Build regression models
regression_results, regression_predictions = build_regression_models(X_train, y_target, feature_names)

print("\n=== REGRESSION RESULTS SUMMARY ===")
print("Predicting 2023 from 2015-2018 data:")
for model_name, model_results in regression_results.items():
    print(f"\n{model_name}:")
    for bloc, metrics in model_results.items():
        print(f"  {bloc}: R² = {metrics['cv_r2_mean']:.3f}, MAE = {metrics['mae']:.2f}%")
```

```python
# Step 17: Analyze feature importance from Random Forest

def analyze_feature_importance(results, feature_names):
    # Use Random Forest for feature importance analysis
    rf_results = results['Random Forest']

    fig, axes = plt.subplots(1, 3, figsize=(20, 6))

    for i, (bloc, bloc_results) in enumerate(rf_results.items()):
        # Get feature importance
        pipeline = bloc_results['pipeline']
        model = pipeline.named_steps['model']

        # Create feature importance dataframe
        importance_df = pd.DataFrame({
            'feature': feature_names,
            'importance': model.feature_importances_
        }).sort_values('importance', ascending=False)

        # Plot top 10 features
        top_features = importance_df.head(10)

        axes[i].barh(range(len(top_features)), top_features['importance'])
        axes[i].set_yticks(range(len(top_features)))
        axes[i].set_yticklabels(top_features['feature'], fontsize=10)
        axes[i].set_xlabel('Feature Importance')
        axes[i].set_title(f'Top 10 Features - {bloc} Bloc')
        axes[i].invert_yaxis()

    plt.tight_layout()
    plt.show()

    # Print top 5 features for each bloc
    print("\n=== TOP 5 MOST IMPORTANT FEATURES ===")
    for bloc, bloc_results in rf_results.items():
        pipeline = bloc_results['pipeline']
        model = pipeline.named_steps['model']

        importance_df = pd.DataFrame({
            'feature': feature_names,
            'importance': model.feature_importances_
        }).sort_values('importance', ascending=False)

        print(f"\n{bloc} Bloc:")
        for idx, row in importance_df.head(5).iterrows():
            print(f"  {row['feature']}: {row['importance']:.3f}")

analyze_feature_importance(regression_results, feature_names)
```

```python
# Step 18: Visualize predictions vs actual results

def create_prediction_visualizations(predictions, y_target):
    # Use the best model (Ridge Regression for consistency)
    best_predictions = predictions['Ridge Regression']

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    blocs = ['Conservative_Right', 'Centrist', 'Left']
    colors = ['#f39c12', '#e74c3c', '#2ecc71']

    for i, (bloc, color) in enumerate(zip(blocs, colors)):
        actual = y_target.iloc[:, i]
        predicted = best_predictions[bloc]

        axes[i].scatter(actual, predicted, alpha=0.6, color=color, s=50)

        # Add perfect prediction line
        min_val = min(actual.min(), predicted.min())
        max_val = max(actual.max(), predicted.max())
        axes[i].plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, alpha=0.8)

        axes[i].set_xlabel(f'Actual {bloc.replace("_", " ")} Support (%)')
        axes[i].set_ylabel(f'Predicted {bloc.replace("_", " ")} Support (%)')
        axes[i].set_title(f'{bloc.replace("_", " ")} Bloc: Predicted vs Actual')
        axes[i].grid(True, alpha=0.3)

        # Add R² score
        r2 = regression_results['Ridge Regression'][bloc]['cv_r2_mean']
        axes[i].text(0.05, 0.95, f'R² = {r2:.3f}', transform=axes[i].transAxes,
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", alpha=0.8))

    plt.tight_layout()
    plt.show()

create_prediction_visualizations(regression_predictions, y_target)
```

```python
# Step 19: Build classification models to predict winning bloc

def prepare_classification_data(df):
    # Create classification target (which bloc won in 2023)
    target_cols = ['Conservative_Right_Bloc_2023', 'Centrist_Bloc_2023', 'Left_Bloc_2023']
    df['Winning_Bloc_2023'] = df[target_cols].idxmax(axis=1)
    df['Winning_Bloc_2023'] = df['Winning_Bloc_2023'].str.replace('_Bloc_2023', '')

    print(f"Classification target distribution:")
    print(df['Winning_Bloc_2023'].value_counts())

    return df['Winning_Bloc_2023']

def build_classification_models(X, y_class):
    models = {
        'Random Forest Classifier': RandomForestClassifier(
            n_estimators=100,
            max_depth=6,
            min_samples_split=3,
            random_state=42
        ),
        'SVM Classifier': SVC(kernel='rbf', C=1.0, random_state=42, probability=True)
    }

    results = {}

    for name, model in models.items():
        print(f"\nTraining {name}...")

        # Create pipeline
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('model', model)
        ])

        # Cross-validation
        cv_scores = cross_val_score(pipeline, X, y_class, cv=5, scoring='accuracy')

        # Fit on all data
        pipeline.fit(X, y_class)

        # Get predictions for confusion matrix
        y_pred = pipeline.predict(X)

        results[name] = {
            'cv_accuracy_mean': cv_scores.mean(),
            'cv_accuracy_std': cv_scores.std(),
            'pipeline': pipeline,
            'predictions': y_pred
        }

        print(f"  Accuracy: {cv_scores.mean():.3f} (±{cv_scores.std():.3f})")

    return results

# Prepare and build classification models
y_classification = prepare_classification_data(final_df)
classification_results = build_classification_models(X_train, y_classification)

# Display classification results with confusion matrices
print("\n=== CLASSIFICATION RESULTS ===")
for model_name, metrics in classification_results.items():
    print(f"\n{model_name}:")
    print(f"  Accuracy: {metrics['cv_accuracy_mean']:.3f} (±{metrics['cv_accuracy_std']:.3f})")

    # Show confusion matrix
    y_pred = metrics['predictions']
    cm = confusion_matrix(y_classification, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Conservative_Right', 'Centrist', 'Left'],
                yticklabels=['Conservative_Right', 'Centrist', 'Left'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

    # Classification report
    print(f"\nClassification Report for {model_name}:")
    print(classification_report(y_classification, y_pred))
```
